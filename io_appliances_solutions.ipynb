6. Handle missing values: Fill missing values with the mean

```python
# Separate numeric columns from non-numeric columns
numeric_cols = data.select_dtypes(include=['float64', 'int64']).columns
non_numeric_cols = data.select_dtypes(exclude=['float64', 'int64']).columns

# Compute the mean only for numeric columns
numeric_means = data[numeric_cols].mean()

# Fill missing values in numeric columns with their respective means
data.loc[:, numeric_cols] = data.loc[:, numeric_cols].fillna(numeric_means)

# Optionally, fill missing values in non-numeric columns (e.g., with a placeholder or mode)
# Example: Fill missing values in non-numeric columns with a placeholder (e.g., 'Unknown')
data.loc[:, non_numeric_cols] = data.loc[:, non_numeric_cols].fillna('Unknown')

# Display a few rows to verify
data.head()
```
